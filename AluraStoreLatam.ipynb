### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



### **En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.**

```python
tienda.shape
```

```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda
```

```python
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda2
```

```python

ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda3
```

```python

ingreso_tienda4 = tienda4['Precio'].sum()
ingreso_tienda4
```

```python
ingresos_Tiendas = {
                    "Tienda1":ingreso_tienda,
                    "Tienda2":ingreso_tienda2,
                    "Tienda3":ingreso_tienda3,
                    "Tienda4":ingreso_tienda4,
                    }
```

```python
df_ingresos_Tiendas = pd.DataFrame(list(ingresos_Tiendas.items()), columns=["Tienda", "Ingresos_Totales"])
df_ingresos_Tiendas.set_index("Tienda", inplace=True)
```

# 2. Ventas por categoría

### **En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.**

```python
ventas_categoria_1 = tienda.groupby('Categoría del Producto')['Producto'].size().sort_values(ascending=False)
ventas_categoria_1
```

```python
ventas_categoria_2 = tienda2.groupby('Categoría del Producto')['Producto'].size().sort_values(ascending=False)
ventas_categoria_2
```

```python
ventas_categoria_3 = tienda3.groupby('Categoría del Producto')['Producto'].size().sort_values(ascending=False)
ventas_categoria_3
```

```python
ventas_categoria_4 = tienda4.groupby('Categoría del Producto')['Producto'].size().sort_values(ascending=False)
ventas_categoria_4
```

```python
ventasxcategoria_Tiendas = {
                    "Tienda1":ventas_categoria_1,
                    "Tienda2":ventas_categoria_2,
                    "Tienda3":ventas_categoria_3,
                    "Tienda4":ventas_categoria_4,
                    }
```

```python
df_ventasxcategoria_Tiendas = pd.DataFrame(list(ventasxcategoria_Tiendas.items()), columns=["Tienda", "Ventas_Categoria"])
df_ventasxcategoria_Tiendas.set_index("Tienda", inplace=True)
```

# 4. Productos más y menos vendidos

### **En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.**

```python

def masVendidos (datos):
  productos_mas_vendidos = datos.groupby('Producto').size().sort_values(ascending=False)
  print(productos_mas_vendidos.head(5))
```

```python
masVendidos1 = masVendidos(tienda)
masVendidos1
```

```python
masVendidos2 = masVendidos(tienda2)
masVendidos2
```

```python
masVendidos3 = masVendidos(tienda3)
masVendidos3
```

```python
masVendidos4 = masVendidos(tienda4)
masVendidos4
```

```python
mas_vendidos = {
                  "Tienda1":masVendidos1,
                  "Tienda2":masVendidos2,
                  "Tienda3":masVendidos3,
                  "Tienda4":masVendidos4,
                        }
```

```python
def menosVendidos (datos):
  productos_menos_vendidos = datos.groupby('Producto').size().sort_values(ascending=True)
  print(productos_menos_vendidos.head(5))
```

```python
menosVendidos1 = menosVendidos(tienda)
menosVendidos1
```

```python
menosVendidos2 = menosVendidos(tienda2)
menosVendidos2
```

```python
menosVendidos3 = menosVendidos(tienda3)
menosVendidos3
```

```python
menosVendidos4 = menosVendidos(tienda4)
menosVendidos4
```

```python
menos_vendidos = {
                  "Tienda1":menosVendidos1,
                  "Tienda2":menosVendidos2,
                  "Tienda3":menosVendidos3,
                  "Tienda4":menosVendidos4,
                        }
```

# 5. Envío promedio por tienda
